# Q1: What are the key limitations of traditional MapReduce that make Spark a preferred choice for modern big data processing?

*  MapReduce stores data on disk after every stage, which increases
processing time.

* It becomes slow when the same data needs to be processed multiple times.

* Writing MapReduce programs can be complex and time-consuming.

* It is not well suited for real-time data analysis.

* Spark is preferred because it provides faster and more efficient processing.




# Q2: Explain how Spark uses In-Memory Computing to speed up iterative machine learning algorithms compared to disk-based systems.
* Spark keeps frequently used data in memory instead of reading it from disk every time.
* This helps machine learning algorithms run faster during multiple iterations.
* Less disk access means less waiting time and better performance.
* Spark can cache datasets and reuse them whenever needed.
* For example, while training a recommendation system, Spark can quickly access the same data from memory and complete the task faster.




# Q3: Write a code snippet to remove all duplicate rows from a DataFrame based on a specific set of columns: user_id and transaction_date.

```python
df_clean = df.dropDuplicates(["user_id", "transaction_date"])
#check if it drops or not
print("Rows after dataset cleaned: ",df_clean.count())
df_clean.show(5)
```

# Q4: Given a DataFrame df_sales, write a query to filter for rows where the region is 'West' and then group by product_category to find the average sale_amount.

```python
#filter
df_result = df.filter(F.col("region") == "West") \
                    .groupBy("product_category") \
                    .agg(F.avg("sale_amount").alias("avg_sale_amount"))

df_result.show()
```

# Q5: What is the difference between .na.drop() and .na.fill()? Provide a code example of filling null values in a status column with the string 'Unknown'.
* **na.drop()** removes rows that contain null or missing values from the DataFrame.
* **na.fill()** replaces null or missing values with a specified value instead of removing the rows.
* **Example:**
If a student's age is missing:
* Using na.drop() will remove that student's record.
 * Using na.fill(0) will keep the record and replace the missing age with 0


```python
df_filled = df.na.fill({"status": "Unknown"})
print (df_filled.show(5))
```

# Q6: Write a query to find the total count of records for each city in a DataFrame, but only for cities where the count is greater than 100.

```python
df_city_count = df.groupBy("city") \
                   .count() \
                   .filter(F.col("count") > 100)
print(df_city_count.count())
df_city_count.show(5)
# as in the dataset there is no city with count >100, so lets see with count >10
df_city_count = df.groupBy("city") \
                   .count() \
                   .filter(F.col("count") > 10)
print(df_city_count.count())
df_city_count.show(5)
```

# Q7: How does the immutability of Spark DataFrames affect how you perform "data cleaning" steps like dropping columns or renaming them?
*  Spark DataFrames are immutable, which means the original DataFrame cannot be changed after it is created.
*  When performing data cleaning tasks like dropping or renaming columns, Spark creates a new DataFrame with the changes instead of modifying the existing one.
*  This helps maintain data consistency and prevents accidental changes to the original data.
*  Example, after using drop() or withColumnRenamed(), we usually store the result in a new DataFrame variable.

*   Example= df_new = df.drop("City")
*  Here, df remains unchanged, and df_new contains the updated data.



# Q8: Write a Spark command to filter a dataset for rows where the age is between 18 and 30 (inclusive) and the subscription is 'Premium'.

```python
df_filtered = df.filter(F.col("age").between(18, 30) & (F.col("subscription") == "Premium"))
print(df_filtered.count())
df_filtered.show(5)
```

# Q9: When cleaning a dataset, why is it often better to handle null values before performing mathematical aggregations like sum() or avg()?

*   Null values can lead to inaccurate or misleading aggregation results if they are not handled properly.
*   Cleaning null values beforehand ensures that calculations such as sum() and avg() are based on complete and meaningful data.
*   It helps improve the overall quality and reliability of data analysis.
*   Depending on the situation, null values can be removed or replaced with suitable values before performing calculations.
*  Example:
If some employees have missing salary values, calculating the average salary without handling those null values may not accurately represent the dataset.

# Q10: Write the code to revise a column named raw_timestamp by casting it to a TimestampType and renaming it to event_time.

```python
df_updated = df.withColumn("event_time", F.col("raw_timestamp").cast(TimestampType())) \
               .drop("raw_timestamp")
print(df_updated.count())
df_updated.show(5)
```

# Q11: Explain the "Shuffle" process that occurs during a grouping operation. Why is it considered a wide transformation?


*   A shuffle occurs when Spark redistributes data across different partitions to bring similar values together for operations like groupBy().
*   During this process, data is transferred between executors, which can be time-consuming.
*   It is called a wide transformation because data from multiple partitions is required to create the new partitions.
*   Shuffling is one of the most expensive operations in Spark due to network and disk usage.
*  Eg: df.groupBy("department").count().show()

# Q12: Write a code snippet that identifies and removes rows where the email column contains null values OR the username is an empty string.


```python
df_cleaned = df.filter(F.col("email").isNotNull() & (F.col("username") != ""))
print(df_cleaned.count())
df_cleaned.show(5)
```

# Q13: How do you use the .agg() function to calculate multiple statistics at once, such as the min, max, and mean of the price column?


```python
df_stats_datas = df.agg(
    F.min("price").alias("min_price"),
    F.max("price").alias("max_price"),
    F.mean("price").alias("mean_price")
)
print(df_stats_datas.count())
df_stats_datas.show(5)
```

# Q14: In the context of cleaning a dataset, what is the risk of using inferSchema=true when your source data contains messy or inconsistent date formats?
*  When inferSchema=true is used, Spark automatically determines the data type of each column based on the values it reads.
*  If a date column contains inconsistent formats, Spark may incorrectly identify the column's data type or convert some values to null.
*  This can lead to data quality issues and incorrect analysis results.
*  Therefore, it is often safer to define the schema manually when working with messy datasets.




# Q15: Write a final processing pipeline that:

Filters out duplicates.

Fills null prices with 0.

Groups by store_id to calculate total revenue.

```python
final_revenue_df = df.dropDuplicates() \
                     .fillna({"price": 0}) \
                     .groupBy("store_id") \
                     .agg(F.sum("price").alias("total_revenue"))

print(final_revenue_df.count())
final_revenue_df.show(20)
```
