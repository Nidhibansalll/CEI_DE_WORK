## Q1: Explain the roles of the Driver, Cluster Manager, and Executor in a Spark application.

1. Driver
*   Acts as the brain of the Spark application and starts the job.
*   Breaks the job into tasks and assigns them to Executors.
*   Collects the results from Executors and returns the final output.
2. Cluster Manager
*   Manages the cluster's resources like CPU and memory.
*   Starts Executors on available worker nodes.
*   Ensures resources are shared efficiently among applications.
3. Executor
*   Executes the tasks assigned by the Driver.
*   Processes data in parallel and caches it in memory when needed.
*   Sends the processed results back to the Driver.**bold text**



# Q2: How does Spark’s Lazy Evaluation strategy improve performance when chain-processing large datasets?

*  Spark doesn't execute transformations immediately. Instead, it records them and creates an execution plan (DAG).
*  Before execution, Spark optimizes the plan by removing unnecessary operations and combining multiple transformations wherever possible.
*  This reduces data movement, minimizes computation, and improves performance, especially when processing large datasets.

# Q4: What is the difference between CSV and Parquet in terms of storage (row-based vs. columnar) and why does it matter for performance?


1.  CSV
* Row-based storage, where data is stored row by row.
* Takes more storage space because it doesn't support compression efficiently.
* Slower for analytics.
2.  Parquet
* Columnar storage, where data is stored column by column.
* Uses compression and encoding, so it requires less storage.
* Faster for analytics.

# Q6: Write the code to "revise" a DataFrame by renaming the column old_name to new_name and casting the price column from a String to a Double.

```python
from pyspark.sql.functions import col

df = df.withColumnRenamed("old_name", "new_name") \
       .withColumn("price", col("price").cast("double"))

print("DataFrame after renaming and casting:")
df.show()
```

# Q7: How does Spark use the Lineage Graph (DAG) to provide fault tolerance if a worker node fails?
*   Spark keeps track of all the transformations performed on the data using a Lineage Graph (DAG).
*   If an Executor or worker node fails, Spark refers to this graph to find what data was lost.
*   It then recomputes only the missing data instead of running the entire job again, which saves both time and resources.



# Q8: Write a query to filter a DataFrame df_orders for rows where the status is 'Completed' AND the amount is greater than 1000.

```python
from pyspark.sql.functions import col

df.filter(
    (col("status") == "Completed") &
    (col("amount") > 1000)
).show()
```

# Q9: Explain the concept of Predicate Pushdown in Parquet and how it affects the amount of data loaded into memory.
*   Predicate Pushdown means Spark applies the filter condition while reading the Parquet file, instead of loading all the data first.
*   It reads only the required rows and columns, skipping the unnecessary data.
*   This reduces the amount of data loaded into memory, making queries faster and more memory-efficient.



# Q10: Write a code snippet to add a new column final_price which is the base_price multiplied by 1.18 (18% tax).

# Q11: What is the difference between Transformations and Actions? Provide two examples of each.

1.  Transformations
*   Transformations create a new DataFrame or RDD without executing the operation immediately.
*   Spark waits until an action is called.
*   Examples: filter(), select()
2.  Actions
*   Actions trigger the execution of all the transformations.
*   They either return a result or write data to storage.
*   Examples: show(), count()



# Q12: Write the Spark command to load a Parquet file from "path/to/input", filter out any rows where user_id is null, and save the result as a CSV at "path/to/output".

```python
#create the input parquet file first
df.write.mode("overwrite").parquet("path/to/input")

df_input = spark.read.parquet("path/to/input")

df_filtered = df_input.filter("user_id IS NOT NULL")

print("Row count after removing nulls:", df_filtered.count())
df_filtered.show(5)

df_filtered.write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv("path/to/output")

print("Saved cleaned data to path/to/output")
```

# Q13: In Spark Architecture, what is the difference between Client Mode and Cluster Mode?
1.   Client Mode
*   The Driver runs on the local machine from where the application is submitted.
*   Good for development, testing, and debugging.
*   If the client machine disconnects or crashes, the Spark application may stop.
2.   Cluster Mode
*   The Driver runs inside the cluster on one of the worker nodes.
*   Suitable for production because the application continues running even if the client disconnects.
*   Provides better reliability and resource management.

# Q14: Write a query to filter a dataset for rows where the region is 'North' OR the priority is 'High'.

# Q15: When exploring a dataset, why is it safer to use .show(5) instead of .collect() on a multi-terabyte dataset?

*   .show(5) displays only the first 5 rows, so it uses very little memory.
*   .collect() loads the entire dataset into the Driver's memory.
*   On a multi-terabyte dataset, using .collect() can lead to memory issues or even crash the application, whereas .show(5) is safe for quickly inspecting the data.
